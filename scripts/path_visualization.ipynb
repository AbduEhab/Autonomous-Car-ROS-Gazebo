#### importing dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open('../path/waypoints.json') as f:
    waypoints = json.load(f)["waypoints"]
```

```python
with open('../path/vehicle_path_file.json') as f:
    vehicle_path = json.load(f)["vehicle_path"]
```

```python
waypoints=np.array(waypoints)*20
vehicle_path=np.array(vehicle_path)
```

```python
plt.figure(figsize=(10,10))
plt.plot(waypoints[:,0], waypoints[:,1], 'r', linewidth=1.5,label='waypoints')
plt.plot(vehicle_path[:,0], vehicle_path[:,1], 'b', linewidth=1.5,label='vehicle_path')
plt.title('Path Visualization')
plt.legend()
plt.show()
```

#### mean square curvature error calculation

```python
total_distance = 0
for x,y in zip(waypoints[:,0], waypoints[:,1]):
    min_distance = float('inf')
    for x1,y1 in zip(vehicle_path[:,0], vehicle_path[:,1]):
        distance = np.sqrt((x-x1)**2 + (y-y1)**2)
        if distance < min_distance:
            min_distance = distance
    total_distance += min_distance**2
total_distance/=len(waypoints)
print("mean square distance: ", total_distance)
```
